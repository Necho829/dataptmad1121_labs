```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris_flower = pd.read_csv('./iris-data.csv')
iris_flower
```

```python
iris_flower.isnull().sum()
```

```python
iris_flower = iris_flower.fillna(iris_flower["petal_width_cm"].median())
iris_flower
```

```python
iris_flower['class'] = iris_flower['class'].replace('Iris-setossa','Iris-setosa')
iris_flower['class'] = iris_flower['class'].replace('versicolor','Iris-versicolor')
iris_flower
```

```python
iris_flower = pd.get_dummies(iris_flower)
iris_flower
```

```python
iris_flower.plot(kind= "scatter", x = "sepal_width_cm",y = "sepal_length_cm", color = "brown")
```

```python
iris_flower.plot(kind= "scatter", x = "petal_width_cm",y = "petal_length_cm", color = "green")
```

```python
iris_flower.plot(kind= "scatter", x = "sepal_length_cm",y = "class_Iris-versicolor", color = "red")
```
